```python
METRIC_COLS = [
    "diversity",
    "temporal_coverage",
    "context_precision",
    "faithfulness_report",
    "answer_relevancy_report",
]

sns.set_theme(style="whitegrid", palette="muted")
```

## 1. Розподіл метрик по типу period

```python
fig, axes = plt.subplots(1, len(METRIC_COLS), figsize=(20, 5))
fig.suptitle("Метрики по типу period (raw_generation)", fontsize=14)

for ax, metric in zip(axes, METRIC_COLS):
    sns.boxplot(data=df, x="period_type", y=metric, order=["week", "month", "year"], ax=ax)
    ax.set_title(metric)
    ax.set_xlabel("")

plt.tight_layout()
plt.show()
```

## 2. Середні метрики по юзерах

```python
user_means = df.groupby("user_id")[METRIC_COLS].mean()

fig, axes = plt.subplots(1, len(METRIC_COLS), figsize=(20, 4))
fig.suptitle("Середні метрики по юзерах (raw_generation)", fontsize=14)

for ax, metric in zip(axes, METRIC_COLS):
    user_means[metric].plot(kind="bar", ax=ax, color=sns.color_palette("muted"))
    ax.set_title(metric)
    ax.set_xlabel("user_id")
    ax.tick_params(axis="x", rotation=0)

plt.tight_layout()
plt.show()
```

## 3. Scatter: faithfulness vs answer_relevancy

```python
scatter_df = df.dropna(subset=["faithfulness_report", "answer_relevancy_report"])

fig, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(
    data=scatter_df,
    x="faithfulness_report",
    y="answer_relevancy_report",
    hue="period_type",
    style="period_type",
    s=80,
    ax=ax,
)
ax.set_title("Faithfulness vs Answer Relevancy (raw_generation)")
ax.set_xlabel("faithfulness_report")
ax.set_ylabel("answer_relevancy_report")
plt.tight_layout()
plt.show()
```

## 4. Heatmap кореляцій між метриками

```python
corr = df[METRIC_COLS].corr()

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True,
    ax=ax,
)
ax.set_title("Кореляція метрик (raw_generation)")
plt.tight_layout()
plt.show()
```

## 5. Tokens та generation time

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Технічні метрики (raw_generation)", fontsize=14)

sns.histplot(data=df, x="report_tokens_used", hue="period_type", kde=True, ax=axes[0])
axes[0].set_title("Розподіл токенів")
axes[0].set_xlabel("tokens")

sns.histplot(data=df, x="report_generation_time", hue="period_type", kde=True, ax=axes[1])
axes[1].set_title("Розподіл часу генерації (сек)")
axes[1].set_xlabel("seconds")

plt.tight_layout()
plt.show()
```
